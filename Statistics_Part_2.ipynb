```python
Q1.What is hypothesis testing in statistics?
Ans- 
   Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. 
   It helps determine whether a certain assumption (hypothesis) about a population parameter is true or not.

### Steps in Hypothesis Testing:
    1. Set Hypotheses:
                      Null Hypothesis (H₀): No effect or no difference.
                      Alternative Hypothesis (H₁): There is an effect or difference.

    2. Pick Significance Level (α): Usually 0.05 (5%).

    3. Choose a Test: t-test, z-test, etc.

    4. Calculate Test Statistic & p-value.

    5. Make a Decision:
                        If p-value ≤ α, reject H₀ (significant result).
                        If p-value > α, fail to reject H₀ (no significant result).
```

```python
Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
Ans-
    The null hypothesis (H₀) states that there is no effect, no difference, or no relationship in the population. 
    It represents the default assumption.

    The alternative hypothesis (H₁ or Ha) is the opposite of H₀ and suggests that there is an effect, a difference, or a relationship in the population.

    ## Key Differences:
                        Feature	                   Null Hypothesis (H₀)	                                Alternative Hypothesis (H₁)
                        Meaning	                   No effect or difference	                            There is an effect or difference
                        Assumption                 Default claim (status quo)	                        Claim being tested
                        Example (New Drug Study)   "The drug has no effect."	                        "The drug improves health."
                        Decision in Testing	       Can be rejected or not rejected	                    Supported if H₀ is rejected

    #In hypothesis testing, we collect data to determine whether to reject H₀ in favor of H₁.
```

```python
Q3. What is the significance level in hypothesis testing, and why is it important?

Ans- The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis (H₀) when it is actually true. 
     It helps determine how much risk we are willing to take for making a wrong decision.

    ## Common Significance Levels:
                                0.05 (5%) → There is a 5% chance of mistakenly rejecting H₀.
                                0.01 (1%) → Stricter test with only a 1% chance of error.

    ## Why is it important?
       It controls the risk of Type I error (rejecting a true H₀).
       Helps decide whether a result is statistically significant or just due to chance.

    # If p-value ≤ α, we reject H₀ and say the result is significant! 
```

```python
Q4. What does a P-value represent in hypothesis testing?

Ans- The P-value in hypothesis testing represents the probability of getting the observed results (or more extreme ones) 
     if the null hypothesis (H₀) is true.

    ## What It Means:
       Small P-value (≤ α, like 0.05) → Strong evidence against H₀ → Reject H₀ (significant result).
       Large P-value (> α) → Weak evidence against H₀ → Fail to reject H₀ (not significant).

    Example:
            If a new drug is tested and the P-value is 0.03, it means there is only a 3% chance that the observed effect is due to random chance. 
            Since 0.03 is less than α = 0.05, we reject H₀ and conclude the drug likely works!
```

```python
Q5. How do you interpret the P-value in hypothesis testing?

Ans- The P-value tells us how likely it is to get our observed results (or something more extreme) if the null hypothesis (H₀) is true.

     ## How to Interpret:
        Small P-value (≤ 0.05) → Strong evidence against H₀ → Reject H₀ (the result is statistically significant).
        Large P-value (> 0.05) → Weak evidence against H₀ → Fail to reject H₀ (not statistically significant).

     Example:
             If a P-value is 0.02 (which is less than 0.05), we reject H₀ and say the result is significant. 
             This means the observed effect is unlikely due to chance
```

```python
Q6. What are Type 1 and Type 2 errors in hypothesis testing?

Ans- In hypothesis testing, Type I and Type II errors are mistakes we can make when making decisions about the null hypothesis (H₀).

    ## Type I Error (False Positive):
    
    1. Rejecting H₀ when it is actually true.
    2. It’s like a false alarm—saying there is an effect when there isn’t.
    3. The probability of making this error is the significance level (α) (e.g., 5%).
       Example: A test says a new drug works, but in reality, it doesn’t.

    ## Type II Error (False Negative):

    1. Failing to reject H₀ when it is actually false.
    2. It’s like missing something real—saying there is no effect when there actually is.
    3. The probability of making this error is β (beta), and 1 - β is the test's power.
       Example: A test says a new drug doesn’t work, but in reality, it does.

    ## Summary:
               Type I error: False positive (wrongly rejecting H₀).
               Type II error: False negative (wrongly accepting H₀).
```

```python
Q7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans- The difference between a one-tailed and a two-tailed test in hypothesis testing lies in how we test the direction of the effect.

    ## One-Tailed Test:
       Tests for an effect in only one direction (either higher or lower, but not both).
       Example: Testing if a new drug is better than the old one (one direction).
    
       Hypotheses:
                  H₀: The new drug is not better.
                  H₁: The new drug is better.
    Used when you have a specific expectation about the direction of the effect.

    ### Two-Tailed Test:
        Tests for an effect in both directions (higher or lower).
        Example: Testing if a new drug is different (either better or worse) from the old one.
    
        Hypotheses:
                    H₀: The new drug has no effect.
                    H₁: The new drug is different (could be better or worse).
    Used when you're unsure of the direction of the effect and want to test both possibilities.

### Summary:

            One-tailed: Checks for a specific direction (e.g., better, higher).
            Two-tailed: Checks for any difference, in either direction.

A one-tailed test is more focused but can miss effects in the opposite direction, while a two-tailed test is broader and more cautious. 
```

```python
Q17. What is the Chi-square distribution, and when is it used?

Ans-
    ## What is the Chi-Square Distribution?
    The Chi-Square (χ²) distribution is used in statistics to analyze categorical data (data in groups like "male/female" or "yes/no").
    It helps check if there’s a real relationship between variables or if differences happened by chance.

    ## When Do We Use It?
    1. To check if data fits an expected pattern (Goodness-of-Fit Test)
       Example: Is a dice fair? (Are all numbers appearing equally?)

    2. To check if two categories are related (Test for Independence)
       Example: Does gender affect coffee vs. tea preference?

    3. To compare different groups (Test for Homogeneity)
       Example: Do people in two different cities have different voting habits?

### How Does It Work?
    We compare actual (observed) counts with expected counts using this formula:
                        χ2 = ∑(O−E)^2 / E

    Where:

    O = Actual number (Observed)
    E = Expected number (if there was no difference)

    If the χ² value is big, it means there's a real difference and not just randomness!

## Example:

Let’s say we survey 100 people about their favorite drink:
	Tea	Coffee	Total
Male	30	20	50
Female	10	40	50
Total	40	60	100

If there’s no real preference, we’d expect equal proportions in both genders. Using the formula, we calculate χ² and see if the difference is big 
enough to matter or just random.
```

```python
Q18. What is the Chi-square goodness of fit test, and how is it applied?

Ans- 
    # What is the Chi-Square Goodness of Fit Test?
        The Chi-Square Goodness of Fit Test is a statistical test used to determine if a sample data matches an expected distribution.
        It helps you assess whether the observed frequencies (actual data) in different categories are consistent with what you would expect based on some hypothesis.

    ## When is it Used?
       Used when you have categorical data (like "Yes/No", "Male/Female", etc.).
       Compares actual data to expected data.
       Example: Are the numbers of people in different age groups equally distributed in a population? Or, are the outcomes of a dice roll as expected?

### How is the Chi-Square Goodness of Fit Test Applied?
    Steps:
     1. State the Hypotheses:
        Null Hypothesis (H₀): The observed data fits the expected distribution (i.e., no difference).
        Alternative Hypothesis (H₁): The observed data does not fit the expected distribution.

     2. Calculate the Expected Frequencies:
        If you expect the categories to be equally distributed, you can calculate the expected frequency by dividing the total number by the number of categories.
        Example: If you're rolling a fair die and you roll it 60 times, you'd expect 10 rolls per number (since there are 6 possible outcomes).

     3. Apply the Chi-Square Formula:
                        χ2=∑(O−E)^2 / E
   
        Where:
        O = Observed frequency (the actual number of occurrences in each category).
        E = Expected frequency (how many occurrences you would expect in each category if the null hypothesis were true).

    4. Calculate the Chi-Square Statistic:
        Once you have the observed and expected values, plug them into the formula to get the χ² value.

    5. Compare with Critical Value:
        Find the critical value from the Chi-Square table using your degrees of freedom (df) and your significance level (usually 0.05).
        Degrees of Freedom (df) = Number of categories - 1.
        If your χ² value is greater than the critical value from the table, reject the null hypothesis.

```

```python
Q19. What is the F-distribution, and when is it used in hypothesis testing?

Ans- 
    # What is the F-distribution?
      The F-distribution is a probability distribution used primarily in variance analysis and comparison of variances.
      It is skewed to the right and is used in tests where we compare the ratio of variances from two or more groups. The F-distribution is defined 
      by two degrees of freedom:
                -> df₁ (numerator degrees of freedom)
                -> df₂ (denominator degrees of freedom)

    The F-statistic is calculated as the ratio of two sample variances (or mean square values) and follows an F-distribution under the null hypothesis.

 ## When is the F-distribution Used in Hypothesis Testing?
    The F-distribution is typically used in the following situations:

    1. Analysis of Variance (ANOVA):
        ANOVA is used to compare the variances between three or more groups to check if there are any statistically significant differences between the means of the groups.
        Example: Comparing exam scores of students from three different teaching methods.

    2. Testing the Equality of Two Variances (F-test):
        The F-test is used to compare the variances of two populations to determine if they are significantly different.
        Example: Checking if the variability of height in two different countries is similar.

## How is the F-distribution Used in Hypothesis Testing?

    1. State the Hypotheses:
        Null Hypothesis (H₀): There is no significant difference between the variances (or means).
        Alternative Hypothesis (H₁): There is a significant difference between the variances (or means).

    2. Calculate the F-statistic:
        For ANOVA, the F-statistic is the ratio of variance between groups to the variance within groups.
                    F = Variance between groups / Variance within groups
        For the two-sample F-test, it's the ratio of the two sample variances:
   
                    F=Sample Variance 1 / Sample Variance 2
   

    3. Find the Critical Value:
        Using the degrees of freedom (df₁ and df₂) and the significance level (usually 0.05), find the critical value from the F-distribution table.

   4.  Make a Decision:
        If the calculated F-statistic is greater than the critical value, reject the null hypothesis (there’s a significant difference).
        If the calculated F-statistic is less than the critical value, fail to reject the null hypothesis (there’s no significant difference).

```

```python
Q20.  What is an ANOVA test, and what are its assumptions?

Ans- 
    # What is an ANOVA Test?
      ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to see if at least one of them is
      significantly different from the others. It helps you test the null hypothesis that the means of all groups are equal, and determines if
      there are any statistically significant differences.
          
    ## When is ANOVA Used?
       Comparing means across multiple groups.
        Example: Comparing the average test scores of students from three different teaching methods.
       When you have a continuous dependent variable and a categorical independent variable with three or more levels (groups).

   ### How Does ANOVA Work?

    1. State the Hypotheses:
        Null Hypothesis (H₀): The means of all groups are equal (no significant difference).
        Alternative Hypothesis (H₁): At least one group mean is different.

    2. Calculate the F-statistic:
        The F-statistic in ANOVA is the ratio of the variance between groups to the variance within groups.
                    F = Variance between groups / Variance within groups
   
        Variance between groups reflects how much the group means differ from the overall mean.
        Variance within groups reflects the variation within each group.

    3. Find the Critical Value:
        Using the degrees of freedom (df) for between-group and within-group variance, and the significance level (usually 0.05), find the critical F-value from the F-distribution table.

    4. Decision Rule:
        If the calculated F-statistic is greater than the critical F-value, reject the null hypothesis.
        If the calculated F-statistic is less than the critical value, fail to reject the null hypothesis.

### Assumptions of the ANOVA Test:
    For ANOVA to produce reliable results, the following assumptions need to be met:

     1. Independence of Observations:
        The data points within each group and between groups should be independent of each other. This means that one group's data should not 
        influence another's.

     2. Normality:
        The data in each group should be approximately normally distributed. You can test this assumption using a normality test
        (like the Shapiro-Wilk test) or visually with histograms/Q-Q plots.

     3. Homogeneity of Variances (Homoscedasticity):
        The variances in each group should be approximately equal. You can test this assumption using a test like Levene's test or Bartlett’s test.
```

```python
Q21.  What are the different types of ANOVA tests?

Ans- 
   ## Different Types of ANOVA Tests

There are several types of ANOVA tests, each designed to handle different study designs and research questions. The most common types of ANOVA are:
1. One-Way ANOVA

    Purpose: Compares the means of three or more independent groups based on one factor.
    Example: Comparing the average test scores of students taught using three different teaching methods.

When to Use:

    You have one independent variable (factor) with three or more levels (groups).
    Each group represents a different treatment or condition.

Assumptions:

    Independence of observations.
    Normality of data in each group.
    Homogeneity of variances (equal variances across groups).

2. Two-Way ANOVA

    Purpose: Compares the means of groups based on two factors (independent variables), and can also assess if there’s an interaction effect between the two factors.
    Example: Comparing the average test scores of students based on teaching method (online, in-person) and gender (male, female).

When to Use:

    You have two independent variables and want to see if they have an effect on a dependent variable.
    You also want to test if there's an interaction effect between the two factors.

Assumptions:

    Same as one-way ANOVA but with two factors.
    Independence of observations.
    Normality and homogeneity of variances for each combination of factors.

3. Repeated Measures ANOVA

    Purpose: Compares the means of three or more related groups based on one factor, where the same subjects are used in all conditions (e.g., measuring the same individuals at different times or under different conditions).
    Example: Measuring the blood pressure of individuals at three different time points: before, during, and after a treatment.

When to Use:

    You have one factor and repeated measurements on the same participants (within-subject design).

Assumptions:

    The dependent variable is normally distributed.
    The variances of the differences between measurements are the same (sphericity).

4. Multivariate Analysis of Variance (MANOVA)

    Purpose: Extends ANOVA to analyze multiple dependent variables simultaneously.
    Example: Measuring the height and weight of individuals from three different diets (three independent groups).

When to Use:

    You have multiple dependent variables (e.g., height and weight) and want to test the effect of one or more factors on them.

Assumptions:

    The data for each dependent variable is multivariate normally distributed.
    The variances and covariances between the dependent variables are equal.

5. Factorial ANOVA

    Purpose: A type of two-way ANOVA, but applied to designs where you have more than two factors, and you test the main effects of each factor as well as their interactions.
    Example: A study looking at how age, gender, and exercise frequency affect body fat percentage.

When to Use:

    When you have more than two factors and need to study their main effects and interaction effects.

Assumptions:

    Same assumptions as for two-way ANOVA but extended to multiple factors.

6. Analysis of Covariance (ANCOVA)

    Purpose: Combines ANOVA and regression to compare one or more group means while controlling for one or more continuous covariates (variables that might influence the dependent variable).
    Example: Comparing the test scores of students in different teaching methods, while controlling for their prior academic performance (covariate).

When to Use:

    When you want to compare group means but control for continuous variables that could affect the dependent variable.

Assumptions:

    Normality and homogeneity of variances for the dependent variable.
    The relationship between the covariate and the dependent variable is linear. 
```

```python
Q22. What is the F-test, and how does it relate to hypothesis testing?

Ans- 
    ## What is the F-test?
       The F-test is a statistical test used to compare the variances of two or more groups. It specifically tests if the variances of two populations
       are significantly different. The F-test is based on the F-distribution, which is the ratio of two sample variances. It's used in various 
       hypothesis tests, including ANOVA, regression analysis, and comparisons of variances.

  ##How is the F-test Used in Hypothesis Testing?

The F-test compares the ratio of two variances to determine if they are significantly different. In hypothesis testing, the F-test is used to:

    Compare Two Population Variances:
        Null Hypothesis (H₀): The two population variances are equal.
        Alternative Hypothesis (H₁): The two population variances are not equal.

    Used in ANOVA:
        The F-test is central to ANOVA (Analysis of Variance), where it compares the variance between group means to the variance within groups to assess whether there are significant differences between group means.

## How Does the F-test Work?

The formula for calculating the F-statistic is:
F=Variance of Group 1 (between-group variance) / Variance of Group 2 (within-group variance)


Here’s how it works:

    Calculate the variances:
        You compute the variance for each group or sample.
        For comparing two groups, you calculate the sample variances for both groups.
    Form the F-statistic:
        The F-statistic is the ratio of the larger variance (between-group variance or group variance) to the smaller variance (within-group variance or error variance). This ratio follows an F-distribution.
    Compare the F-statistic to the critical value:
        Using degrees of freedom for both the numerator (between groups) and the denominator (within groups), you look up the critical value of F from an F-distribution table at a specific significance level (usually 0.05).
    Decision Rule:
        If the calculated F-statistic is greater than the critical value, reject the null hypothesis.
        If the calculated F-statistic is less than the critical value, fail to reject the null hypothesis.

### Relationship Between F-test and Hypothesis Testing

    The F-test is used in hypothesis testing to compare variances between two or more groups.
    In ANOVA, the F-test compares the variability between group means to the variability within groups to determine if there are significant differences in group means.
    The F-statistic provides the test statistic in these tests, helping you decide whether to reject or fail to reject the null hypothesis based on the ratio of variances.
```

```python
#answer-
import numpy as np
from scipy import stats

# Sample data (example)
sample_data = [51, 53, 52, 50, 49, 51, 54, 52, 53, 51]  # Sample of 10 people
population_mean = 50  # Known population mean
population_std = 2  # Known population standard deviation

# Step 1: Calculate the sample mean and sample size
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 2: Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 3: Determine the critical value for a two-tailed test at alpha = 0.05
alpha = 0.05  # Significance level
critical_value = stats.norm.ppf(1 - alpha / 2)  # For two-tailed test

# Step 4: Perform the Z-test and interpret the results
if abs(z_statistic) > critical_value:
    result = "Reject the null hypothesis. There is a significant difference between the sample mean and population mean."
else:
    result = "Fail to reject the null hypothesis. There is no significant difference between the sample mean and population mean."

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"Critical Value (for alpha = 0.05): {critical_value}")
print(f"Interpretation: {result}")
```

```python
Q5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
```

```python
#answer-
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_visualize_errors(pop_mean, pop_std, alt_mean, sample_size, alpha=0.05):
    # Step 1: Define the null and alternative distributions
    null_distribution = stats.norm(loc=pop_mean, scale=pop_std / np.sqrt(sample_size))
    alt_distribution = stats.norm(loc=alt_mean, scale=pop_std / np.sqrt(sample_size))

    # Step 2: Calculate the critical value for a two-tailed test
    critical_value = stats.norm.ppf(1 - alpha / 2)
    
    # Step 3: Calculate Type I Error (area under the null distribution in the rejection region)
    type_I_error = 1 - stats.norm.cdf(critical_value, loc=pop_mean, scale=pop_std / np.sqrt(sample_size))
    
    # Step 4: Calculate Type II Error (area under the alternative distribution in the non-rejection region)
    type_II_error = stats.norm.cdf(critical_value, loc=alt_mean, scale=pop_std / np.sqrt(sample_size))

    # Step 5: Plot the distributions and visualize errors
    x = np.linspace(pop_mean - 4 * pop_std, pop_mean + 4 * pop_std, 1000)

    # Plot the null and alternative distributions
    plt.plot(x, null_distribution.pdf(x), label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, alt_distribution.pdf(x), label='Alternative Hypothesis (H₁)', color='orange')
    
    # Shade Type I error region (rejection region for H₀)
    plt.fill_between(x, 0, null_distribution.pdf(x), where=(x > critical_value), color='red', alpha=0.5, label='Type I Error')
    
    # Shade Type II error region (non-rejection region for H₀ but under H₁)
    plt.fill_between(x, 0, alt_distribution.pdf(x), where=(x < critical_value), color='green', alpha=0.5, label='Type II Error')
    
    # Mark critical value on the plot
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
    
    # Labels and title
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Test Statistic (Z)')
    plt.ylabel('Probability Density')
    plt.legend()
    
    # Show the plot
    plt.show()

    # Return the calculated errors
    return type_I_error, type_II_error

# Example usage of the function
pop_mean = 50  # Population mean (null hypothesis)
pop_std = 10   # Population standard deviation
alt_mean = 55  # Alternative hypothesis mean (true mean)
sample_size = 30  # Sample size

type_I, type_II = calculate_and_visualize_errors(pop_mean, pop_std, alt_mean, sample_size)

print(f"Type I Error: {type_I:.4f}")
print(f"Type II Error: {type_II:.4f}")
```

```python
Q9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
```

```python
#answer-
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Function to calculate the confidence interval for a sample mean.

    Parameters:
    - sample_data: Array or list of sample data
    - confidence_level: Desired confidence level (default is 0.95)

    Returns:
    - A tuple (lower_bound, upper_bound) representing the confidence interval.
    """
    # Step 1: Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(sample_data)

    # Step 2: Calculate the standard error (SE)
    standard_error = sample_std / np.sqrt(sample_size)

    # Step 3: Determine the critical value from the t-distribution
    # (Using t-distribution since it's typically used for smaller sample sizes)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Step 4: Calculate the margin of error (ME)
    margin_of_error = critical_value * standard_error

    # Step 5: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulating sample data

# Calculate the 95% confidence interval for the sample mean
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
```

```python
#answer-
def bayesian_inference(prior_h, likelihood_h, likelihood_not_h, prior_not_h):
    """
    Function to perform Bayesian Inference using Bayes' Theorem.
    
    Parameters:
    - prior_h: Prior probability of hypothesis (P(H)), e.g., probability of having the disease.
    - likelihood_h: Likelihood of evidence given hypothesis (P(E|H)), e.g., probability of testing positive given the disease.
    - likelihood_not_h: Likelihood of evidence given not hypothesis (P(E|¬H)), e.g., probability of testing positive without the disease (false positive rate).
    - prior_not_h: Prior probability of not hypothesis (P(¬H)), e.g., probability of not having the disease.
    
    Returns:
    - Posterior probability (P(H|E)): The updated probability of the hypothesis after observing the evidence.
    """
    # Calculate the evidence (P(E))
    evidence = (likelihood_h * prior_h) + (likelihood_not_h * prior_not_h)
    
    # Apply Bayes' Theorem to compute the posterior probability
    posterior_h = (likelihood_h * prior_h) / evidence
    
    return posterior_h

# Given probabilities
prior_h = 0.01          # P(H): 1% prior probability of having the disease
likelihood_h = 0.9      # P(E|H): 90% chance of testing positive given the disease
likelihood_not_h = 0.05 # P(E|¬H): 5% false positive rate
prior_not_h = 0.99      # P(¬H): 99% prior probability of not having the disease

# Perform Bayesian Inference
posterior = bayesian_inference(prior_h, likelihood_h, likelihood_not_h, prior_not_h)

print(f"The posterior probability of having the disease given a positive test result is: {posterior:.4f}")
```

```python
#answer-
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Function to calculate the expected frequencies for a Chi-square test.
    
    Parameters:
    - observed_data: 2D numpy array representing the observed frequencies (contingency table).
    
    Returns:
    - expected_frequencies: 2D numpy array of expected frequencies.
    """
    # Step 1: Compute row totals
    row_totals = np.sum(observed_data, axis=1)
    
    # Step 2: Compute column totals
    column_totals = np.sum(observed_data, axis=0)
    
    # Step 3: Compute the grand total
    grand_total = np.sum(observed_data)
    
    # Step 4: Calculate expected frequencies
    expected_frequencies = np.outer(row_totals, column_totals) / grand_total
    
    return expected_frequencies

# Example contingency table (observed data)
observed_data = np.array([[30, 10, 20],  # Male preferences (Action, Drama, Comedy)
                          [25, 15, 30]]) # Female preferences (Action, Drama, Comedy)

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Data:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
Q16. Implement an F-test using Python to compare the variances of two random samples.
```

```python
#answer-
import numpy as np
import scipy.stats as stats

def perform_f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two samples.
    
    Parameters:
    - sample1: First sample data (array-like).
    - sample2: Second sample data (array-like).
    
    Returns:
    - F-statistic: The calculated F-test statistic.
    - p-value: The p-value of the F-test.
    """
    # Calculate the variances of the samples
    var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance for sample2
    
    # Determine the larger and smaller variance
    if var1 > var2:
        F_statistic = var1 / var2
        df1 = len(sample1) - 1
        df2 = len(sample2) - 1
    else:
        F_statistic = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1
    
    # Calculate the p-value using the F-distribution
    p_value = 1 - stats.f.cdf(F_statistic, df1, df2)
    
    return F_statistic, p_value

# Example: Generate two random samples with different variances
np.random.seed(42)  # For reproducibility

sample1 = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std=10, n=100
sample2 = np.random.normal(loc=50, scale=5, size=100)   # Mean=50, Std=5, n=100

# Perform the F-test
F_statistic, p_value = perform_f_test(sample1, sample2)

print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

# Interpretation of the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#answer-
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def perform_anova(*groups):
    """
    Perform a one-way ANOVA test to compare the means of multiple groups.
    
    Parameters:
    - *groups: Multiple groups of data to compare (array-like).
    
    Returns:
    - F-statistic: The calculated F-statistic for the ANOVA test.
    - p-value: The p-value of the ANOVA test.
    """
    # Perform the ANOVA test
    F_statistic, p_value = stats.f_oneway(*groups)
    
    return F_statistic, p_value

# Example: Generate data for three groups
np.random.seed(42)  # For reproducibility

group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Perform the ANOVA test
F_statistic, p_value = perform_anova(group1, group2, group3)

# Output the results
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

# Interpretation of the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")

# Visualization of the data
plt.boxplot([group1, group2, group3], labels=["Group 1", "Group 2", "Group 3"])
plt.title("ANOVA Test: Comparison of Group Means")
plt.ylabel("Values")
plt.show()
```

```python
Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
```

```python
#answer-
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    """
    Check the assumptions for a one-way ANOVA test:
    1. Normality of each group
    2. Equal variances (homogeneity of variances)
    
    Parameters:
    - *groups: Multiple groups of data (array-like).
    
    Returns:
    - Assumptions check results (normality p-value, equal variance p-value).
    """
    
    # 1. Check normality assumption using Shapiro-Wilk test for each group
    normality_results = {}
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        normality_results[f"Group {i} Normality (Shapiro-Wilk)"] = p_value
    
    # 2. Check equal variance assumption using Levene's test
    _, p_value_levene = stats.levene(*groups)
    
    return normality_results, p_value_levene

def plot_data(*groups):
    """
    Plot the boxplot of the data for visual inspection of normality and equal variance.
    
    Parameters:
    - *groups: Multiple groups of data (array-like).
    """
    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, labels=[f"Group {i+1}" for i in range(len(groups))])
    plt.title("Boxplot of Groups for Assumptions Check")
    plt.ylabel("Values")
    plt.show()

# Example: Generate data for three groups
np.random.seed(42)  # For reproducibility

group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Check the assumptions for ANOVA
normality_results, p_value_levene = check_anova_assumptions(group1, group2, group3)

# Output normality results for each group
for group, p_value in normality_results.items():
    print(f"{group}: p-value = {p_value}")
    
# Output Levene's test for equal variances
print(f"\nLevene's Test for Equal Variance: p-value = {p_value_levene}")

# Interpretation of results
alpha = 0.05
print("\nInterpretation of Assumptions Check:")
for group, p_value in normality_results.items():
    if p_value < alpha:
        print(f"{group}: Reject null hypothesis (Data is not normally distributed).")
    else:
        print(f"{group}: Fail to reject null hypothesis (Data is normally distributed).")

if p_value_levene < alpha:
    print("Levene's Test: Reject null hypothesis (Variances are not equal).")
else:
    print("Levene's Test: Fail to reject null hypothesis (Variances are equal).")

# Visualize the data to check assumptions visually
plot_data(group1, group2, group3)
```

```python
#answer-
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Step 1: Simulate data
np.random.seed(42)

# Create two factors (Factor 1: "Treatment" and Factor 2: "Time")
treatments = ['A', 'B']
times = ['Morning', 'Afternoon']

# Generate data for each combination of factors
data = []
for treatment in treatments:
    for time in times:
        # Simulating a continuous outcome (e.g., weight) with some random noise
        outcome = np.random.normal(loc=50 + (5 * (treatments.index(treatment) - 0.5)) + (3 * (times.index(time) - 0.5)), scale=10, size=30)
        data.append([treatment, time, outcome])

# Create a DataFrame
df = pd.DataFrame(data, columns=["Treatment", "Time", "Outcome"])

# Step 2: Perform Two-Way ANOVA
model = ols('Outcome ~ C(Treatment) + C(Time) + C(Treatment):C(Time)', data=df).fit()
anova_results = anova_lm(model)

# Output the ANOVA table
print(anova_results)

# Step 3: Interpretation of Results
alpha = 0.05
print("\nInterpretation of Results:")
if anova_results['PR(>F)'][0] < alpha:
    print("There is a significant effect of Treatment.")
else:
    print("There is no significant effect of Treatment.")
    
if anova_results['PR(>F)'][1] < alpha:
    print("There is a significant effect of Time.")
else:
    print("There is no significant effect of Time.")
    
if anova_results['PR(>F)'][2] < alpha:
    print("There is a significant interaction between Treatment and Time.")
else:
    print("There is no significant interaction between Treatment and Time.")

# Step 4: Visualization of interaction effect
plt.figure(figsize=(8, 6))
sns.boxplot(x='Treatment', y='Outcome', hue='Time', data=df)
plt.title("Two-Way ANOVA: Interaction between Treatment and Time")
plt.show()
```

```python
#answer-
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data for a sample
np.random.seed(42)

# Simulate data: Population variance = 100, sample size = 30
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Step 2: Calculate sample variance
sample_variance = np.var(sample_data, ddof=1)

# Step 3: Set up hypotheses
sigma_0_squared = 100  # Hypothesized population variance (sigma^2 = 100)
n = len(sample_data)  # Sample size
df = n - 1  # Degrees of freedom

# Step 4: Calculate the test statistic
chi_square_statistic = (n - 1) * sample_variance / sigma_0_squared

# Step 5: Find the critical value for two-tailed test
alpha = 0.05  # Significance level
lower_critical_value = stats.chi2.ppf(alpha / 2, df)
upper_critical_value = stats.chi2.ppf(1 - alpha / 2, df)

# Step 6: Interpret results
print(f"Chi-square Statistic: {chi_square_statistic}")
print(f"Lower Critical Value: {lower_critical_value}")
print(f"Upper Critical Value: {upper_critical_value}")

# Step 7: Conclusion
if chi_square_statistic < lower_critical_value or chi_square_statistic > upper_critical_value:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

```python
Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
```

```python
#answer-
import numpy as np
import scipy.stats as stats

# Step 1: Define the data
# Group 1: 100 people, 60 successes
# Group 2: 120 people, 75 successes
n1 = 100  # Sample size for Group 1
x1 = 60   # Successes in Group 1

n2 = 120  # Sample size for Group 2
x2 = 75   # Successes in Group 2

# Step 2: Calculate the sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Step 3: Calculate the pooled sample proportion
p = (x1 + x2) / (n1 + n2)

# Step 4: Calculate the standard error
se = np.sqrt(p * (1 - p) * (1 / n1 + 1 / n2))

# Step 5: Calculate the Z-statistic
z_statistic = (p1 - p2) / se

# Step 6: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

# Step 7: Output the results
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Step 8: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```
